```python
import rasterio

# Open the input TIFF file
with rasterio.open('/content/drive/MyDrive/Spring_2019/Sentinel_Laikipia_County_Spring_19-0000000000-0000000000.tif') as src:
    # Read the image data
    image = src.read()

    # Replace NaN values with 0
    image = image.copy()
    image[np.isnan(image)] = 0

    # Update metadata for output file
    kwargs = src.meta
    kwargs.update(dtype=rasterio.float32, nodata=0)

    # Create the output TIFF file
    with rasterio.open('/content/drive/MyDrive/Spring_2019/output.tif', 'w', **kwargs) as dst:
        # Write the modified image data to the output file
        dst.write(image)
```
