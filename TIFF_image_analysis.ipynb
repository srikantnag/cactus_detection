```python
import rasterio
from rasterio.plot import show
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
img=rasterio.open('/content/drive/MyDrive/Sentinel_Laikipia_County.tif')
```

```python
img.shape
```

```python
full_img=img.read()
```

```python
full_img.shape
```

```python
img_band4 = img.read(4)
img_band1 = img.read(1) #1 stands for 1st band. 
img_band2 = img.read(2) #2 stands for 2nd band. 
img_band16 = img.read(16) #3 stands for 3rd band. 
img_band8 = img.read(8) #3 stands for 3rd band. 
img_band3 = img.read(3)
img_band5 = img.read(5)
img_band6 = img.read(6)
img_band7 = img.read(7)
img_band9 = img.read(9)
img_band10 = img.read(10)
img_band11 = img.read(11)
img_band12 = img.read(12)
img_band13 = img.read(13)
img_band14 = img.read(14)
img_band15 = img.read(15)
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(8,2,1)
ax1.imshow(img_band1, cmap='Blues')
ax2 = fig.add_subplot(8,2,2)
ax2.imshow(img_band2, cmap='pink')
ax3 = fig.add_subplot(8,2,3)
ax3.imshow(img_band16, cmap='pink')
ax4 = fig.add_subplot(8,2,4)
ax4.imshow(img_band8, cmap='Greens')
ax5 = fig.add_subplot(8,2,5)
ax5.imshow(img_band4, cmap='pink')
ax6 = fig.add_subplot(8,2,6)
ax6.imshow(img_band3, cmap='Greens')
ax7 = fig.add_subplot(8,2,7)
ax7.imshow(img_band5, cmap='pink')
ax8 = fig.add_subplot(8,2,8)
ax8.imshow(img_band6, cmap='Greens')
ax9 = fig.add_subplot(8,2,9)
ax9.imshow(img_band7, cmap='pink')
ax10 = fig.add_subplot(8,2,10)
ax10.imshow(img_band9, cmap='Greens')
ax11 = fig.add_subplot(8,2,11)
ax11.imshow(img_band10, cmap='pink')
ax12 = fig.add_subplot(8,2,12)
ax12.imshow(img_band11, cmap='Greens')
ax13 = fig.add_subplot(8,2,9)
ax13.imshow(img_band12, cmap='pink')
ax14 = fig.add_subplot(8,2,10)
ax14.imshow(img_band13, cmap='Greens')
ax15 = fig.add_subplot(8,2,11)
ax15.imshow(img_band14, cmap='pink')
ax16 = fig.add_subplot(8,2,12)
ax16.imshow(img_band15, cmap='Greens')
```
